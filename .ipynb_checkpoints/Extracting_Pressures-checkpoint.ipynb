```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import both the data and the metadata
with open("20200912-NSH-ATL_886b2a47-3249-4e95-8200-f7cdd8fbbf46_SecondSpectrum_Metadata.json") as f:
    metadata = json.load(f)

data = []    
with open("20200912-NSH-ATL_886b2a47-3249-4e95-8200-f7cdd8fbbf46_SecondSpectrum_Data.jsonl") as f:
    for line in f:
        data.append(json.loads(line.rstrip('\n|\r')))
        
```

```python
# This will come in handy later
id_to_name = {}
for i in range(len(metadata['homePlayers'])):
    player_id = metadata['homePlayers'][i]['optaId']
    player_name = metadata['homePlayers'][i]['name']
    id_to_name[player_id] = [player_name, "Home"]
    
for i in range(len(metadata['awayPlayers'])):
    player_id = metadata['awayPlayers'][i]['optaId']
    player_name = metadata['awayPlayers'][i]['name']
    id_to_name[player_id] = [player_name, "Away"]
id_to_name
```

```python
# As pitchLength and pitchWidths are in meters, I'm going to assume that the
# coordinate system is in meters, which means I'm going to have to convert meters to yards,
# as the prompt was to find pressure situations
# (defending team player within 5 yards of the player in possession of the ball and moving towards the ball


data_backup = data  # save to a backup just in case

# First, check to make sure there is 22 players on the field in every frame
for frame in range(len(data)):
    if (len(data[frame]['homePlayers']) != 11):
        print("Not 11 Home Players - frame", frame)
    if (len(data[frame]['awayPlayers']) != 11):
        print("Not 11 Away Players - frame", frame)
        
# No output, so can be sure there is 22 players on field in every frame
```

```python
# Now to get to converting
# 1 meter = 1.09361 yards

# Assume of data[frame]['ball']['xyz'] is None, the ball is out of play

for frame in range(len(data)):
    for i in range(11):
        data[frame]['homePlayers'][i]['xyz'][0] = data[frame]['homePlayers'][i]['xyz'][0] * 1.09361
        data[frame]['homePlayers'][i]['xyz'][1] = data[frame]['homePlayers'][i]['xyz'][1] * 1.09361
        data[frame]['homePlayers'][i]['xyz'][2] = data[frame]['homePlayers'][i]['xyz'][2] * 1.09361
        data[frame]['homePlayers'][i]['speed'] = data[frame]['homePlayers'][i]['speed'] * 1.09361
        
        data[frame]['awayPlayers'][i]['xyz'][0] = data[frame]['awayPlayers'][i]['xyz'][0] * 1.09361
        data[frame]['awayPlayers'][i]['xyz'][1] = data[frame]['awayPlayers'][i]['xyz'][1] * 1.09361
        data[frame]['awayPlayers'][i]['xyz'][2] = data[frame]['awayPlayers'][i]['xyz'][2] * 1.09361
        data[frame]['awayPlayers'][i]['speed'] = data[frame]['awayPlayers'][i]['speed'] * 1.09361
    if (data[frame]['ball']['xyz'] is not None):
        data[frame]['ball']['xyz'][0] = data[frame]['ball']['xyz'][0] * 1.09361
        data[frame]['ball']['xyz'][1] = data[frame]['ball']['xyz'][1] * 1.09361
        data[frame]['ball']['xyz'][2] = data[frame]['ball']['xyz'][2] * 1.09361
        data[frame]['ball']['speed'] = data[frame]['ball']['speed'] * 1.09361
```

```python
# Now we're going to get to work finding all the frames in which a player is in possession
# We'll define that as having the ball within a yard for at least a full second (25 frames)

possession_data = pd.DataFrame(columns=['frame', 'player_in_poss', 'player_in_poss2'], index=range(len(data)))

# We want to to start on the 25th frame - a full second has passed.
idx = 0 #this needs to reach 25 to count as a player in possession

#for frame in range(25,50):
for frame in range(25,len(data)):
    if (data[frame]['ball']['xyz'] is not None): # ball in play
        ball_x = data[frame]['ball']['xyz'][0]
        ball_y = data[frame]['ball']['xyz'][1]
        ball_z = data[frame]['ball']['xyz'][2]
        
        players_inside_yard = []
        players_inside_yard_dist = []
        for i in range(11):
            player_x = data[frame]['homePlayers'][i]['xyz'][0]
            player_y = data[frame]['homePlayers'][i]['xyz'][1]
            player_z = data[frame]['homePlayers'][i]['xyz'][2]
            
            # 2D or 3D
            #dist_from_ball = np.sqrt(np.power((ball_x-player_x),2) + np.power((ball_y-player_y),2) + np.power((ball_z-player_z),2))
            dist_from_ball = np.sqrt(np.power((ball_x-player_x),2) + np.power((ball_y-player_y),2))
            if (dist_from_ball < 1):
                players_inside_yard.append(data[frame]['homePlayers'][i]['playerId'])
                players_inside_yard_dist.append(dist_from_ball) # KEEP AN EYE
        for i in range(11):
            player_x = data[frame]['awayPlayers'][i]['xyz'][0]
            player_y = data[frame]['awayPlayers'][i]['xyz'][1]
            player_z = data[frame]['awayPlayers'][i]['xyz'][2]
            
            #dist_from_ball = np.sqrt(np.power((ball_x-player_x),2) + np.power((ball_y-player_y),2) + np.power((ball_z-player_z),2))
            dist_from_ball = np.sqrt(np.power((ball_x-player_x),2) + np.power((ball_y-player_y),2))
            if (dist_from_ball < 1):
                players_inside_yard.append(data[frame]['awayPlayers'][i]['playerId'])
                players_inside_yard_dist.append(dist_from_ball)
        
        if (len(players_inside_yard) == 1):
            possession_data['frame'][frame] = frame
            possession_data['player_in_poss'][frame] = players_inside_yard[0]
        elif (len(players_inside_yard) == 2):
            possession_data['frame'][frame] = frame
            if (players_inside_yard_dist[0] < players_inside_yard_dist[1]):
                possession_data['player_in_poss'][frame] = players_inside_yard[0]
                possession_data['player_in_poss2'][frame] = players_inside_yard[1]
            else:
                possession_data['player_in_poss'][frame] = players_inside_yard[1]
                possession_data['player_in_poss2'][frame] = players_inside_yard[0]
            
possession_data = possession_data[~possession_data['frame'].isna()]            
# Now, we've to all frames where at least one player is within a yard of the ball
# Find a series of 25 frames in a row where the same player is listed, and that's a second of possession.
```

```python
# Find the number of situations in which there is at least 25 frames in a row
# Save the possession start frame, possesion end frame, and player on the ball

possession_data = possession_data.reset_index().drop(columns=['index'])
possession_start_end = pd.DataFrame(columns=['startFrame', 'endFrame'], index=range(len(possession_data)))

ctr=0
poss_ctr=0
for i in range(1, len(possession_data)):
    if (possession_data['frame'][i] - possession_data['frame'][i-1] == 1): #current and one before are consecutive
        ctr+=1
    else:
        if (ctr >= 25):
            possession_start_end['startFrame'][i] = possession_data['frame'][i-1-ctr]
            possession_start_end['endFrame'][i] = possession_data['frame'][i-1]
        ctr=0 # if not consecutive, reset counter

possession_start_end = possession_start_end[~possession_start_end['startFrame'].isna()]
possession_start_end = possession_start_end.reset_index().drop(columns=['index'])
```

```python
possession_start_end # Now we have a record of exactly when possessions start and end, to tell who is on the ball
```

```python
# checking to see how long each possession is
possession_lengths = []
for i in range(len(possession_start_end)):
    possession_lengths.append((possession_start_end['endFrame'][i]-possession_start_end['startFrame'][i])*.04)

bins=[0,1,2,3,4,5,6,7]    
plt.hist(possession_lengths, bins=bins)
        
```

```python
# Make sure the same player is on the ball through all of these possessions

mismatch_frames = []
for i in range(len(possession_start_end)):
    startFrame = possession_start_end['startFrame'][i]
    endFrame = possession_start_end['endFrame'][i]
    
    playerid = possession_data[possession_data['frame'] == startFrame].reset_index().drop(columns=['index'])['player_in_poss'][0]
    for frame in range(startFrame, endFrame+1):
        tmp_id = possession_data[possession_data['frame'] == frame].reset_index().drop(columns=['index'])['player_in_poss'][0]
        tmp_id2 = possession_data[possession_data['frame'] == frame].reset_index().drop(columns=['index'])['player_in_poss2'][0]
        if (tmp_id != playerid):
            if (tmp_id2 == playerid):
                print("SWITCHED SPOTS ----", frame)
            else:
                playername = id_to_name[playerid]
                tmpname = id_to_name[tmp_id]
                print("MISMATCH! ", playername, "----", tmpname, "----", frame)
                mismatch_frames.append(frame)
                break
                
mismatch_frames                
```

```python
# Data is not super consistent, so it has to be cleaned
# We want there to be one player in possession for each possession
# For the mismatch situations, we want to cut off the possession where the mismatch occurs
# For the switched spots, we do nothing for now

for mismatch_frame in mismatch_frames:
    for i in range(len(possession_start_end)):
        if ((mismatch_frame > possession_start_end['startFrame'][i]) and (mismatch_frame < possession_start_end['endFrame'][i])):
            if ((mismatch_frame - possession_start_end['startFrame'][i]) < 25): # not a possession if cut to less than a second long now
                possession_start_end = possession_start_end.drop(index=i)
                possession_start_end = possession_start_end.reset_index().drop(columns=['index'])
                print("Too Short Now - possession dropped from dataframe")
                break
            else:
                length = (mismatch_frame - possession_start_end['startFrame'][i])*.04
                possession_start_end['endFrame'][i] = mismatch_frame-1
                print("All Good -", length, "seconds - possession length adjusted")
                break
```

```python
# Now that we have when possessions start and end, we can see when players are within 5 yards,
# and moving toward the ball


# Record the player in possession so that we can omit them from players pressuring
possession_start_end['player_in_poss'] = ""

for i in range(len(possession_start_end)):
    possStart = possession_start_end['startFrame'][i]
    possEnd = possession_start_end['endFrame'][i]
    
    players_on_ball = []
    for frame in range(possStart, possEnd):
        idx = possession_data.index[possession_data['frame'] == frame].tolist()[0]
        player1_id = possession_data['player_in_poss'][idx]
        player2_id = possession_data['player_in_poss2'][idx]
        if (type(player2_id) == str):
            players_on_ball.append(player1_id)
            players_on_ball.append(player2_id)
        else:
            players_on_ball.append(player1_id)
    
    # Found on stackoverflow to find the majority element in a list
    myMap = {}
    maximum = ( '', 0 ) # (occurring element, occurrences)
    for n in players_on_ball:
        if n in myMap:
            myMap[n] += 1
        else:
            myMap[n] = 1
        # Keep track of maximum on the go
        if myMap[n] > maximum[1]: maximum = (n,myMap[n])
    
    player_name = id_to_name[maximum[0]]
    percent_in_poss = (maximum[1] / (possEnd-possStart))*100
    possession_start_end["player_in_poss"][i] = maximum[0]
    print(player_name, " -- ", percent_in_poss)

# Great, for every possession, we didn't have a switch in possession partway through
```

```python
# Now that we have the player in possession, we should see how many unique players are within 5 yards of the ball,
# and moving closer to it during the course of the possession.

pressures_data = pd.DataFrame(columns=['startFrame', 'endFrame', 'player_in_poss', 'players_pressuring'], index=range(len(data)))

ctr=0
for i in range(len(possession_start_end)):
    possStart = possession_start_end['startFrame'][i]
    possEnd = possession_start_end['endFrame'][i]
    
    players_within_five = []
    poss_HomeOrAway = id_to_name[possession_start_end['player_in_poss'][i]][1]
    if (poss_HomeOrAway == "Home"):
        opps = 'awayPlayers'
    else:
        opps = 'homePlayers'
    
    for frame in range(possStart, possEnd):
        ball_x = data[frame]['ball']['xyz'][0]
        ball_y = data[frame]['ball']['xyz'][1]
        
        for player_i in range(11):
            player_id = data[frame][opps][player_i]['playerId'] 
            player_x = data[frame][opps][player_i]['xyz'][0]
            player_y = data[frame][opps][player_i]['xyz'][1]

            dist_from_ball = np.sqrt(np.power(ball_x-player_x,2)+np.power(ball_y-player_y,2))
            
            if ((dist_from_ball < 5) and (player_id not in players_within_five)):
                players_within_five.append(player_id)
    
    name = id_to_name[possession_start_end['player_in_poss'][i]][0]
    names = []
    for k in range(len(players_within_five)):
        names.append(id_to_name[players_within_five[k]][0])
    if (len(players_within_five) > 0):
        print(possStart, "-", possEnd, ": ", len(players_within_five), "opponents within 5 yards: ", name, "vs.", names)
        ctr+=1
        
        pressures_data['startFrame'][possStart] = possStart
        pressures_data['endFrame'][possStart] = possEnd
        pressures_data['player_in_poss'][possStart] = possession_start_end['player_in_poss'][i]
        pressures_data['players_pressuring'][possStart] = players_within_five
        
print(' ')        
print(ctr, "possessions with at least one player within 5 yards")

pressures_data = pressures_data[~pressures_data['startFrame'].isna()]

# I'm really looking for the 7305-7385 result, because I know from the video that is a one-on-one between
# D. Badji and M. Robinson (4:51-ish)
```

```python
pressures_data = pressures_data.reset_index().drop(columns=['index'])
pressures_data
```

```python
# Define pressure as moving toward the ball.
# And moving toward the ball as getting closer to the ball in 10(?) consecutive frames

pressures_data_final = pd.DataFrame(columns=['startFrame', 'wallClock', 'gameClock', 'playerInPossessionId',
                                             'playerInPossessionName', 'defenderId', 'defenderName'],
                                    index=range(len(data)))

for i in range(len(pressures_data)):
    possStart =  pressures_data['startFrame'][i]
    possEnd = pressures_data['endFrame'][i]
    for player_id in pressures_data['players_pressuring'][i]:
        ctr=0
        dist_list = []
        for frame in range(possStart, possEnd):
            ball_x = data[frame]['ball']['xyz'][0]
            ball_y = data[frame]['ball']['xyz'][1]

            if (id_to_name[player_id][1] == "Home"): team = 'homePlayers'
            else: team = 'awayPlayers'

            for k in range(11):
                if (data[frame][team][k]['playerId'] == player_id):
                    break
            player_x = data[frame][team][k]['xyz'][0]
            player_y = data[frame][team][k]['xyz'][1]

            dist_to_ball = np.sqrt(np.power(ball_x-player_x,2)+np.power(ball_y-player_y,2))
            dist_list.append(dist_to_ball)
        for j in range(10, len(dist_list)+1):
            tmp_list = dist_list[j-10:j]
            if ((tmp_list[0] > tmp_list[1]) and (tmp_list[1] > tmp_list[2]) and (tmp_list[2] > tmp_list[3]) and 
                (tmp_list[3] > tmp_list[4]) and (tmp_list[4] > tmp_list[5]) and (tmp_list[5] > tmp_list[6]) and 
                (tmp_list[6] > tmp_list[7]) and (tmp_list[7] > tmp_list[8]) and (tmp_list[8] > tmp_list[9])): # const increasing
                if ((tmp_list[0] < 5) or (tmp_list[1] < 5) or (tmp_list[2] < 5) or 
                    (tmp_list[3] < 5) or (tmp_list[4] < 5) or (tmp_list[5] < 5) or 
                    (tmp_list[6] < 5) or (tmp_list[7] < 5) or (tmp_list[8] < 5) or (tmp_list[9] < 5)): # inside 5yd
                    frame = (j-10) + possStart
                    pressures_data_final['startFrame'][frame] = frame
                    pressures_data_final['wallClock'][frame] = data[frame]['wallClock']
                    pressures_data_final['gameClock'][frame] = data[frame]['gameClock']
                    pressures_data_final['playerInPossessionId'][frame] = pressures_data['player_in_poss'][i]
                    pressures_data_final['playerInPossessionName'][frame] = id_to_name[pressures_data['player_in_poss'][i]][0]
                    pressures_data_final['defenderId'][frame] = player_id
                    pressures_data_final['defenderName'][frame] = id_to_name[player_id][0]
                    break
                
pressures_data_final = pressures_data_final[~pressures_data_final['startFrame'].isna()]
pressures_data_final = pressures_data_final.reset_index().drop(columns='index')
pressures_data_final
            
```
